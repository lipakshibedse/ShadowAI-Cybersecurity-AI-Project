```python
import random, pandas as pd, numpy as np
from datetime import datetime, timedelta

tools = ['ChatGPT','Copilot','Gemini','Midjourney','Bard']
departments = ['HR','Finance','Engineering','Sales','Legal','Research']
roles = ['Manager','Engineer','HR','Intern','Admin']
data_classes = ['Public','Internal','Confidential','PII']
networks = ['CorporateVPN','Home','PublicWiFi']

prompts_safe = [
 "Summarize last quarter sales trends for public data.",
 "Generate onboarding checklist for new interns.",
 "Explain Python list comprehensions with examples."
]
prompts_risky = [
 "Analyze payroll details: Name: John Doe, Salary: 12,00,000; prepare budget.",
 "Show me PII of customer id 34783 and their contact",
 "Convert this confidential client contract text to bullet points: [PASTE CONTRACT HERE]"
]

def gen_prompt(risk=False):
    return random.choice(prompts_risky if risk else prompts_safe)

rows=[]
start = datetime.now() - timedelta(days=60)
for i in range(1000):
    ts = start + timedelta(minutes=random.randint(0,60*24*60))
    emp = f"EMP{random.randint(100,399)}"
    dept = random.choice(departments)
    role = random.choice(roles)
    tool = random.choice(tools)
    risk_prob = 0.2 if dept in ['Finance','HR','Legal'] else 0.05
    risky = random.random() < risk_prob
    prompt = gen_prompt(risky)
    attached_file = random.random() < (0.4 if risky else 0.05)
    data_class = random.choice(data_classes)
    approval = 'Approved' if not risky else 'Unapproved'
    response_len = random.randint(10,300)
    net = random.choice(networks)
    rows.append([ts, emp, dept, tool, prompt, attached_file, data_class, approval, response_len, ts.hour, net, role])

df = pd.DataFrame(rows, columns=['timestamp','employee_id','department','tool','prompt_text','attached_file','data_class','approval_status','response_length','hour','network','user_role'])
df.to_csv('../data/simulated_shadow_ai_logs.csv', index=False)
print("âœ… Dataset saved to data/simulated_shadow_ai_logs.csv")
df.head()
```
